-INTEGER
Целое число, могут принимать

-NUMERIC(6,3)
Вещественное число,

Дата в формате ГГГГ-ММ-ДД
26 июля 2020 года = '2020-07-26'
3 января 2021 года = '2021-01-03'

-VARCHAR(10)
Строка длиной не более 255 символов, 
в скобках максимальная длина строки,
которая может храниться в поле

-INT PRIMARY KEY AUTO_INCREMENT
Для описания ключевого поля 
(каждая следующая строка будет на 1 больше предыдущего)

```python
SELECT name_author, SUM(amount) AS Количество
FROM author LEFT JOIN book
    ON author.author_id = book.author_id
GROUP BY name_author
HAVING Количество < 10 OR Количество IS NULL
ORDER BY Количество;

/* вывести авторов и сумму кол-ва НАЗВАТЬ Количество
из тбл автор с тбл книги
по кл id автора
группируя сумму по автору
кот имеет Количество меньше 10 или Количество нуль
сортируя Количество по возврастанию */

+----------------+------------+
| name_author    | Количество |
+----------------+------------+
| Лермонтов М.Ю. | NULL       |
| Пастернак Б.Л. | 2          |
| Булгаков М.А.  | 8          |
+----------------+------------+
```

```python
SELECT name_author, SUM(amount) as Количество
FROM 
    author INNER JOIN book
    on author.author_id = book.author_id
GROUP BY name_author
HAVING SUM(amount) = 
     (
      SELECT MAX(sum_amount) AS max_sum_amount
      FROM 
          (
            SELECT author_id, SUM(amount) AS sum_amount 
            FROM book GROUP BY author_id
          ) query_in
      );

/* вывести авторов и сумму кол-ва НАЗВАТЬ Количество
из тбл авторы с тбл книги
по кл id автора
группируя сумму по имени автора
кот имеет сумму кол-ва РАВНУЮ максимальной сумме кол-ва книг 
    из тбл  query_in (вывести id автора и сумму кол-ва
                      из тбл книг группируя по id автора)*/

+------------------+------------+
| name_author      | Количество |
+------------------+------------+
| Достоевский Ф.М. | 23         |
+------------------+------------+
```

Для книг, которые уже есть на складе (в таблице book), но по другой цене, чем в поставке (supply),  
необходимо в таблице book увеличить количество на значение, указанное в поставке,  и пересчитать цену. 
А в таблице  supply обнулить количество этих книг. 

Формула для пересчета цены:
( p1 * k1 + p2 * k2) / k1+k2
где  p1, p2 - цена книги в таблицах book и supply;
     k1, k2 - количество книг в таблицах book и supply.

```python
UPDATE book AS b
    INNER JOIN author AS a USING (author_id)
    INNER JOIN supply AS s 
        ON s.title = b.title
        AND s.author = a.name_author
SET b.amount = b.amount + s.amount,
    s.amount = 0,
    b.price = (b.amount * b.price + s.amount * s.price) / (b.amount + s.amount)
WHERE b.price <> s.price;

/* изменить значения из таблицы книг 
с тбл авторы по кл id автора 
с тбл поставка 
по кл название книги
по кл имя атотра
изменить знач: кол-во книг в тбл книг на сумму кол-ва в 2-ух тбл
кол-во книг в поставке = 0
цена в тбл книг = цене по формуле 
если цены не совпадают*/

Query result:
+---------+-----------------------+-----------+----------+--------+--------+
| book_id | title                 | author_id | genre_id | price  | amount |
+---------+-----------------------+-----------+----------+--------+--------+
| 1       | Мастер и Маргарита    | 1         | 1        | 670.99 | 3      |
| 2       | Белая гвардия         | 1         | 1        | 540.50 | 5      |
| 3       | Идиот                 | 2         | 1        | 437.11 | 13     |
| 4       | Братья Карамазовы     | 2         | 1        | 799.01 | 3      |
| 5       | Игрок                 | 2         | 1        | 480.50 | 10     |
| 6       | Стихотворения и поэмы | 3         | 2        | 650.00 | 15     |
| 7       | Черный человек        | 3         | 2        | 570.20 | 6      |
| 8       | Лирика                | 4         | 2        | 518.99 | 2      |
+---------+-----------------------+-----------+----------+--------+--------+
Affected rows: 8
Query result:
+-----------+-----------------------+------------------+--------+--------+
| supply_id | title                 | author           | price  | amount |
+-----------+-----------------------+------------------+--------+--------+
| 1         | Доктор Живаго         | Пастернак Б.Л.   | 380.80 | 4      |
| 2         | Черный человек        | Есенин С.А.      | 570.20 | 6      |
| 3         | Белая гвардия         | Булгаков М.А.    | 540.50 | 7      |
| 4         | Идиот                 | Достоевский Ф.М. | 360.80 | 0      |
| 5         | Стихотворения и поэмы | Лермонтов М.Ю.   | 255.90 | 4      |
| 6         | Остров сокровищ       | Стивенсон Р.Л.   | 599.99 | 5      |
+-----------+-----------------------+------------------+--------+--------+
Affected rows: 6
```

№17.3
Сравнить ежемесячную выручку от продажи книг за текущий и предыдущий годы.
Для этого вывести год, месяц, сумму выручки 
 в отсортированном сначала по возрастанию месяцев, затем по возрастанию лет виде

 Цена книги для текущего года хранится в таблице book, а для предыдущего в buy_archive

```python
SELECT YEAR(date_payment) AS Год,
	   MONTHNAME(date_payment) AS Месяц,
	   SUM(amount*ba.price) AS Сумма
FROM buy_archive ba
GROUP BY 1, 2

UNION ALL
SELECT YEAR(date_step_end) AS Год,
	   MONTHNAME(date_step_end) AS Месяц,
	   SUM(bb.amount*price) AS Сумма

/*1-ый вариант*/
FROM buy_book bb
      JOIN buy_step bs ON bb.buy_id = bs.buy_id 
  				  AND bs.date_step_end 
  				  AND bs.step_id = 1
      JOIN book USING(book_id)

GROUP BY 1, 2
ORDER BY 2, 1;

/*2-ой вариантСравнить ежемесячную выручку от продажи книг за текущий и предыдущий годы.*/
FROM buy_book bb
      JOIN buy_step bs USING (buy_id) 
      JOIN book USING(book_id)
 WHERE bs.date_step_end IS NOT NULL
 AND bs.step_id = 1

GROUP BY YEAR(date_step_end), MONTHNAME(date_step_end)
ORDER BY MONTHNAME(date_step_end), YEAR(date_step_end);

/*вывести год из даты_оплаты
    месяц из даты оплаты
    сумму произведений кол-во * цену(из архива)
из тбл архив_покупок
группировка по году и месяцу

соединяем с дубликатами
выборку год из даты_окончания_этапа
      месяц из даты_окончания_этапа
      сумму произведений кол-во * цену(из тбл купленных_книг)
из тбл купленные_книги 
    с тбл этап_покупки по кл id покупки
                       и
                       и этап покупки = Оплата
    с тбл книги по кл id книги
группируя по году и месяцу
сортируя по месяцу и году по возрастанию*/

+------+----------+---------+
| Год  | Месяц    | Сумма   |
+------+----------+---------+
| 2019 | February | 5626.30 |
| 2020 | February | 3309.37 |
| 2019 | March    | 6857.50 |
| 2020 | March    | 2131.49 |
+------+----------+---------+
```
